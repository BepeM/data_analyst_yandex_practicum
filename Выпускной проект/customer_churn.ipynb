## Предобработка данных

### Приведение данных к корректному типу

```python
# Изменим наименования столбцов
df.rename(columns = {'USERID':'user_id', 'EST_SALARY':'est_salary'}, inplace = True)
```

```python
# Заменим пустые значения в возрасте средним значением по всему DF, преобрауем тип данных
df['age'] = df['age'].fillna(42)
df['age'] = df['age'].astype(int)
df['score'] = df['score'].astype(int)
df['est_salary'] = df['est_salary'].astype(int)
```

### Проверка на наличие дубликатов

```python
# Проверим наличие дубликатов
df.duplicated().sum()
```

```python
# Проверим наличие неявных дубликатов в столбце `user_id`
df['user_id'].duplicated().sum()
```

```python
user_duplicated = df[df['user_id'].duplicated(keep = False)].sort_values('user_id')
user_duplicated
```

```python
user_duplicated['churn'].value_counts()
```

```python
user_duplicated.groupby('city')['user_id'].count().plot(kind='bar', figsize=(15,5), color=['lightblue', 'pink'])
plt.xticks(rotation=45)
plt.title('Количество клиентов с одинаковым id в разных городах')
plt.xlabel('Город')
plt.ylabel('Количество клиентов')
plt.show()
```

При рассмотрение клиентов с дублированным `user_id`, принимаю решиние оставить все как есть. Мы не можем быть уверены во всех случаях, что это одни и те же люди. Кроме того эти клиенты могут быть полезны в составлении портретов клиентов. С чем связана ошибка по уникальным `user_id` не ясна 

### Работа с некорректными значениями

При ознакомлении с таблицей внимание привлекает минимальное значение в столбце products - 0. Если человек является клиентом банка, то значит он пользуется как минимум одним продуктом.

```python
df[df['products']==0]
```

Судя по всему, 0 в столбце products появился потому что этот клиент ушел из банка и теперь действительно больше не пользуется банковскими продуктами. Однако для целей анализа нам необходимо изучать последние актуальные данные на момент, когда отточные пользователи еще являлись клиентами банка, поэтому эту строку необходимо удалить.

```python
# удалим строку с ошибкой:
df = df[df['products']!=0]
```

### Работа с пропусками

```python
# cоздадим копию таблицы data и заменим пропуски на нули в столбце balance
data_copy = df.copy()
data_copy['balance'].replace(np.nan, 0, inplace=True)

# создадим столбец, который будет показывать, что у клиента нет счета (баланс равен 0)
data_copy['account'] = data_copy['balance'].copy() 
data_copy.account = data_copy.account.astype(np.int64)
data_copy.loc[(data_copy['account'] != 0), 'account'] = 1
data_copy.head()

# cоздадим отдельный датафрейм
yes_balance = data_copy[data_copy['account'] == 1]
no_balance = data_copy[data_copy['balance'] == 0]
no_balance.head(5)
```

```python
# Посмотрим в каких городах у нас встречаются пропуски значений баланса на счетах клиентов
round(no_balance['city'].value_counts()/no_balance['city'].count()*100,2)
```

Пропуски видим в трех городах: Ярославль ~ 64%, Рыбинск ~ 26% и Ростов ~ 10% 

```python
# Посмотрим сколько клиентов у нас попало в отток
no_balance['churn'].value_counts()
```

Видим, что у нас 12 клиентов ушли в отток, то есть такие клиенты все-таки в большинстве остаются.

```python
# Теперь посмотрим на маркер "активности" клиентов за последнее время
no_balance['last_activity'].value_counts()
```

Среди таких клиентов у нас те, кто был активен и не активен почти равны.

```python
# Посмотрим наличие кредитной карты у таких клиентов
round(no_balance['credit_card'].value_counts()/no_balance['credit_card'].count()*100,2)
```

```python
df['balance'] = df['balance'].fillna(0)
df['balance'] = df['balance'].astype(int)
```

```python
display(yes_balance.agg({'median', 'mean'}))
display(no_balance.agg({'median', 'mean'}))
```

Видим, что у клиентов без баланса практицески отсутствует оценка собственности, так же в среднем у этих пользователей на несколько пунктов больше пользования кредитной картой, количество продуктов банка так же разница. У клиентов без баланса в среднем в два раза больше поступления на счет, как и кредитный рейтинг больше на несколько пунктов. В основном это клиенты, которых устраивает банк

Видим, что почти у 82% таких клиентов есть кредитная карта. В целом, можем предположить, что действительно система дает некий сбой, который происходит во всех городах, поэтому перед построением моделей мы не будем заполнять пропущенные значения нулевыми, но какую-то ставку на эту переменную в модели делать не будем (все-таки метод довольно радикальный).

### Кодирование переменных

```python
df['gender'] = df['gender'].replace({'Ж':0, 'М': 1})
df['city_2']=df['city'] # создал копию столбца
data = pd.get_dummies(df, columns = ['city_2'] )
data.rename(columns = {'city_2_Ростов':'city_rostov', 'city_2_Рыбинск':'city_rybinsk', 'city_2_Ярославль':'city_yaroslavl'}, inplace = True)
data.head()
```

**Вывод:**

Мы привели названия столбцов в единый вид для удобства работы, изменили буквенное отображение пола клиентов в численный, рассмотрели клиентов у которых пропуски в столбце balance. Так же, перевели столбец баланса и зарплаты в целочисленный тип для дальнейшего удобства работы с этими столбцами. Можем сделать следующие выводы и предположения:

Очень низкие значения зарплат могут являться как ошибкой, так и реальными значениями в некоторых случаях. Учитывая то, что значения зарплат указны точно до сотых, значит данные записываются с транзакций, а так как таких клиентов достаточно много и у большинства есть объекты в собственности, мы можем предположить, что низкие значения зарплаты возникают по причинам, например каак:

  - Клиент оформился на малую ставку и отработал пару дней на новой работе и мы видим сумму зачислений от работадателя. А значит при анализе этих клиентов через месяц-два мы увидим другие значения зарплат.

  - Клиенты получают дивиденты через наш банк, которые отображаются у нас как зарплата.

  - Клиенты получают пассивных доход через наш банк, когда основной заработок проходит через другой банк.

По пропускам в столбце баланса средств на счетах мы можем предположить, что это не пропущенные значения, а техническая "маска" для пустых значений, то есть, когда на счету у клиента ничего нету, нулевые значения указываются как пропуск.

## Исследовательский анализ данных

### Корреляционный анализ

```python
data.drop('user_id', axis= 1 , inplace= True )
data.sample(5)
```

```python
# вызовем метод corr(), чтобы найти коэффициент корреляции для всех столбцов, и построим тепловую карту:

plt.figure(figsize=(16,8))
sns.heatmap (data.corr(), annot=True, vmin=-1, vmax=1, center= 0, cmap= 'coolwarm', linewidths=1);
```

**Вывод:**
- Ярко выраженная (сильная) корреляция между уходом клиентов и другими факторами отсутствует.

- Согласно шкале Чеддока есть умеренная корреляция с equity.

- Между возрастом, зарплатой, городом и уходом клиента почти нет никакой связи.

- При этом существует слабая корреляция с кол-ом продуктов, активностью, наличием кредитки, балансом, полом и кредитным рейтингом.

Проверим себя, применив корреляци Пирсена

```python
plt.figure(figsize = (6,6))
sns.heatmap(data.corr().loc[:,['churn']].round(2).drop('churn').sort_values('churn', ascending=False), annot = True, vmin=-1, vmax=1, center= 0, cmap= 'coolwarm')
plt.show()
```

Картина схожа, хоть и отличается. Существует умеренная корреляция для кол-ва продуктов и собственности, в остальном либо слабая корреляция, либо её отсутствие. Таким образом, прямой и сильной связи между уходом клиента и другими факторами обнаружить не удалось.

Слабая корреляция просматривается по активности, балансу, полу клиента, кредитному рейтингу и наличию кредитной карточки. Так же почти нет никакой связи между возрастом, зарплатой, городом.

Посмотрим выборочное среднее для ушедших и не ушедших пользователей, чтобы понять можем ли мы применить t test

```python
mean_of_data_churn = []
for _ in range(10000):
    mean_of_data_churn.append(data_churn['est_salary'].sample(1000).mean())
plt.rcParams["figure.figsize"] = (8,5)
pd.Series(mean_of_data_churn).hist()
plt.title('Выборочное срденее зарплат')
plt.xlabel('Заработная плата')
plt.ylabel('Количество')
plt.show()
```

```python
mean_of_data_non_churn = []
for _ in range(1000):
    mean_of_data_non_churn.append(data_non_churn['est_salary'].sample(1000).mean())
plt.rcParams["figure.figsize"] = (8,5)
pd.Series(mean_of_data_non_churn).hist()
plt.title('Выборочное срденее зарплат')
plt.xlabel('Заработная плата')
plt.ylabel('Количество')
plt.show()
```

Распределние выборочных средних похоже на нормальное, проверим используя тест Шапиро-Уилка

```python
alpha = 0.05 
results = st.shapiro(mean_of_data_churn)
results1 = st.shapiro(mean_of_data_non_churn)
p_value = results[1]  
p_value1 = results1[1]  
print('p-значение data_churn: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально')
    
print('p-значение data_non_churn: ', p_value1)
if p_value1 < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально')
```

Очевидно, что большие выбросы влияют на выборку в non churn сегменте. Тем не менее будем считать, что требование к нормальности выборок удовлетворено

Видно большое количество "выбросов" на графике зарплат. Максимальные зарплаты сильно отличаются от медианных значений и даже тех, которые относятся к 75% выборки. Эти "выборосы" могут быть следствием ошибок в системе, вранья клиентов, но могут быть и настоящими данными - мы этого проверить не можем. Однако ввиду того, что для банка состоятельные клиенты с высокой зарплатой большая ценность, то мы не будем удалять эту информацию из датасета, так как удаление этих "выбросов" может исказить результаты исследования портретов пользователей.

```python
salary_non_churn = data['est_salary'][data['churn'] == 0]
salary_churn = data['est_salary'][data['churn'] == 1]
```

Будем проводить t тест, так как несмотря на ненормальность самого распределения, выборочные среднии распределены нормально (или близко к тому). Второй момент: мы имеем слишком большие разницы в объеме выборок - Манна Уитни при таких вводных может давать некорректные результаты.

```python
alpha = 0.05 # уровень статистической значимости
results = st.ttest_ind(salary_churn, salary_non_churn, equal_var = False)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя') 
```

Таким образом, нет статистически значимых различий между доходами ушедших клиентов и тех, кто остался для отклонения нулевой гипотезы.

Проверим эту же гипотезу U тестом Манна Уитни:

```python
alpha = 0.05  

results = st.mannwhitneyu(salary_churn, salary_non_churn)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя') 
```

Результаты данного теста указыват, что есть статистичеси значимые различия в зарплате и они ярко выражены. Считаем, что результатам непараметрического теста Манна Уитни нельзя верить в нашем случае, ввиду большой разницы в объеме выборок.
Принимаем результаты t теста.

Будем проводить t-тест на очищенных от части выбросов данных.

```python
not_emissions_salary_non_churn = data.query('est_salary < 350000')['est_salary'][data['churn'] == 0]
not_emissions_salary_churn = data.query('est_salary < 350000')['est_salary'][data['churn'] == 1]
```

```python
alpha = 0.05 # уровень статистической значимости
results = st.ttest_ind(not_emissions_salary_churn , not_emissions_salary_non_churn, equal_var = False)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя') 
```

Таким образом, есть статистически значимые различий между доходами ушедших клиентов и тех, кто остался для отклонения нулевой гипотезы. Принимаем результаты t теста без выбросов.
